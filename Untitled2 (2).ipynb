```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_data= pd.read_csv('/customer behaviour analysis.csv')
```

```python
customer_data.head ()
```

```python
from datetime import date
import pandas as pd
#recency


snapshot_date = pd.to_datetime('2023-09-16')



customer_data['Purchase Date'] = pd.to_datetime(customer_data['Purchase Date'], format='%d-%m-%Y %H:%M')
recency = customer_data.groupby(['Customer ID']).agg({
    'Purchase Date': lambda x: (snapshot_date - x.max()).days
))
recency.rename(columns={'Purchase Date': 'Recency'}, inplace=True)
recency
```

```python
freq_df = customer_data.drop_duplicates(subset=["Customer ID", "Purchase Date"])
freq = freq_df.groupby(["Customer ID"]).size().reset_index(name='Frequency')
```

```python
freq.head()
```

```python
#monetary

customer_data["Total Purchase Amount"] - customer_data["Product Price"] * customer_data["Quantity"]
```

```python
monetary_value = customer_data.groupby("Customer ID")["Total Purchase Amount"].sum()
monetary_value
```

```python
rfm_df = pd.concat([recency, freq, monetary_value], axis=1)
rfm_df.head()
```

```python
rfm_df.columns
```

```python
spend_cols = ('Recency', 'Customer ID', 'Frequency', 'Total Purchase Amount')
```

```python
rfm_df["total_spending"] = rfm_df[list(spend_cols)].sum(axis=1)
rfm_df.head()
```

```python
rfm_df["total_spending"]
```

```python
def segment(row):
    if row['Total Purchase Amount'] > 3000:
        return "High Value"
    elif row['Total Purchase Amount'] > 1000:
        return "Medium Value"
    else:
        return "Low Value"

rfm_df['Segment'] = rfm_df.apply(segment, axis=1)
rfm_df.head()
```

```python
def customer_segment(row):
    if row['Recency'] <= 171 and row['Frequency'] >= 6.0 and row['Total Purchase Amount'] >= 12524:
        return 'Best Customers'
    elif row['Recency'] <= 60 and row['Frequency'] >= 3:
        return 'Loyal Customers'
    elif row['Recency'] <= 30 and row['Frequency'] == 1:
        return 'New Customers'
    elif row['Recency'] > 60 and row['Frequency'] >= 3:
        return 'At Risk Customers'
    else:
        return 'Lost Customers'

rfm_df['Segment'] = rfm_df.apply(customer_segment, axis=1)
```

```python
customer_data['Purchase Date'] = pd.to_datetime(customer_data['Purchase Date'])

daily_sales = customer_data.groupby('Purchase Date')['Total Purchase Amount'].sum()

plt.figure(figsize=(10,5))
daily_sales.plot(kind='line')
plt.title('Daily Purchase Trend')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
plt.show()
```

```python
monthly_sales = customer_data.groupby(customer_data['Purchase Date'].dt.to_period('M'))['Total Purchase Amount'].sum()

monthly_sales.index = monthly_sales.index.astype(str)

plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line')
plt.title('Monthly Purchase Trend')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.show()
```
